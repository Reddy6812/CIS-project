```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import os
import pickle
# Unified model
from transformers import Wav2Vec2Model


# Define the AudioCNN class with adjustments for input size
class AudioCNN(nn.Module):
    def __init__(self, num_classes, input_size):
        super(AudioCNN, self).__init__()
        self.conv1 = nn.Conv1d(1, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv1d(32, 64, kernel_size=3, stride=1, padding=1)
        self.conv4 = nn.Conv1d(64, 128, kernel_size=3, stride=1, padding=1)  # Added layer
        self.conv5 = nn.Conv1d(128, 256, kernel_size=3, stride=1, padding=1) # Added layer
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(256 * (input_size // 32), 128)  
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = F.relu(self.conv3(x))
        x = self.pool(x)
        x = F.relu(self.conv4(x))  # Applying the fourth conv layer
        x = self.pool(x)
        x = F.relu(self.conv5(x))  # Applying the fifth conv layer
        x = self.pool(x)
        x = x.view(x.size(0), -1)  # Flatten the features
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x


class AudioClassifier(nn.Module):
    def __init__(self, wav2vec2_model, num_classes, input_size):
        super(AudioClassifier, self).__init__()
        self.wav2vec2 = wav2vec2_model
        self.audio_cnn = AudioCNN(num_classes, input_size)

    def forward(self, input_values):
        # Ensure you are using the output's last_hidden_state
        features = self.wav2vec2(input_values).last_hidden_state.mean(dim=1)
        logits = self.audio_cnn(features)
        return logits

# Function to load data
def load_data(dataset_path, data_type, sampling_rate=16000):
    data = []
    labels = []
    processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")

    for file_number in range(1, 51):
        filename = f"data_{file_number}_{data_type}.pkl"
        file_path = os.path.join(dataset_path, filename)
        if os.path.exists(file_path):
            with open(file_path, 'rb') as file:
                file_data = pickle.load(file)
                for waveform, label in file_data:
                    data.append(waveform.squeeze().numpy())
                    labels.append(label)

    labels = [x[0] for x in labels]
    return torch.stack([torch.tensor(d) for d in data]), torch.tensor(labels)

# Load and prepare data
device = torch.device("cuda")
dataset_path = 'emi_dataset/'
train_data, train_labels = load_data(dataset_path, "train")
train_loader = DataLoader(TensorDataset(train_data, train_labels), batch_size=32, shuffle=True)

validate_data, validate_labels = load_data(dataset_path, "valid")
validate_loader = DataLoader(TensorDataset(validate_data, validate_labels), batch_size=32)

test_data, test_labels = load_data(dataset_path, "test")
test_loader = DataLoader(TensorDataset(test_data, test_labels), batch_size=32)

# Initialize the model and move it to GPU
num_classes = 10  # Update this based on your dataset
input_size = 1024  # Adjust based on your feature extraction
wav2vec2_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-large-960h").to(device)
audio_classifier = AudioClassifier(wav2vec2_model, num_classes, input_size).to(device)



# Training loop
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(audio_classifier.parameters(), lr=0.001)

num_epochs = 30  # Adjust as needed
for epoch in range(num_epochs):
    audio_classifier.train()
    for batch_idx, (waveforms, labels) in enumerate(train_loader):
        # Convert waveforms to float
        waveforms = waveforms.float().to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = audio_classifier(waveforms)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        if batch_idx % 10 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Step {batch_idx+1}/{len(train_loader)}, Loss: {loss.item():.4f}")


    # Validation loop
    audio_classifier.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    for waveforms, labels in validate_loader:
        waveforms = waveforms.float().to(device)
        labels = labels.to(device)
        outputs = audio_classifier(waveforms)
        loss = criterion(outputs, labels)
        val_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    val_loss /= len(validate_loader)
    print(f'Epoch {epoch+1}, Validation Loss: {val_loss:.4f}, Validation Accuracy: {(100 * correct / total):.2f}%')

# Test loop
audio_classifier.eval()
test_loss = 0.0
correct = 0
total = 0
for waveforms, labels in test_loader:
    waveforms = waveforms.float().to(device)
    labels = labels.to(device)
    outputs = audio_classifier(waveforms)
    loss = criterion(outputs, labels)
    test_loss += loss.item()
    _, predicted = torch.max(outputs.data, 1)
    total += labels.size(0)
    correct += (predicted == labels).sum().item()
test_loss /= len(test_loader)
print(f'Test Loss: {test_loss:.4f}, Test Accuracy: {(100 * correct / total):.2f}%')
```
